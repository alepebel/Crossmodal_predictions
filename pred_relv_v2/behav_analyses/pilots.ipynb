```python
def accuracyV(df):
    hits = len(df[(df["v_pred"] == "EXP") & (df["resp"] == "familiar")]) + len(df[(df["v_pred"] == "VP") & (df["resp"] == "unfamiliar")])
    misses = len(df[(df["v_pred"] == "EXP") & (df["resp"] != "familiar")]) + len(df[(df["v_pred"] == "VP") & (df["resp"] != "familiar")])
    
    acc = (hits/(hits+misses)) *100
    
    return acc

def accuracyA(df):
    hits = len(df[(df["a_pred"] == "EXP") & (df["resp"] == "familiar")]) + len(df[(df["a_pred"] == "VP") & (df["resp"] == "unfamiliar")])
    misses = len(df[(df["a_pred"] == "EXP") & (df["resp"] != "familiar")]) + len(df[(df["a_pred"] == "VP") & (df["resp"] != "familiar")])
    
    acc = (hits/(hits+misses)) *100
    
    return acc   
```

```python
def block_accuracy(df, modality):
    if modality == "visual":
        hits = len(df[(df.iloc[:,8] == "EXP") & (df.iloc[:,15] == "familiar")]) + \
               len(df[(df.iloc[:,8] == "VP") & (df.iloc[:,15] == "unfamiliar")]) + \
               len(df[(df.iloc[:,8] == "N") & (df.iloc[:,15] == "unfamiliar")]) + \
               len(df[(df.iloc[:,8] == 1) & (df.iloc[:,15] == "catch")])
    else:
        hits = len(df[(df.iloc[:,11] == "EXP") & (df.iloc[:,15] == "familiar")]) + \
               len(df[(df.iloc[:,11] == "VP") & (df.iloc[:,15] == "unfamiliar")]) + \
               len(df[(df.iloc[:,11] == "N") & (df.iloc[:,15] == "unfamiliar")]) + \
               len(df[(df.iloc[:,11] == 1) & (df.iloc[:,15] == "catch")])
                  
    accuracy = str(round(((hits / len(df)) * 100), 2))    
    return accuracy
```

```python
for block in [0,2,4,6]:
    accuracy = block_accuracy(df[df["block"]==block], "auditory")
    print(accuracy)
```
